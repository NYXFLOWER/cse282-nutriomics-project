```python
import pandas as pd
import numpy as np
```

```python
nim_files = [
    "./nim-aminoacidsD_400.csv",
    "./nim-aminoacids_400.csv",
    "./nim-sugars_400.csv",
    "./nim-vitamins_400.csv",
]
```

```python
df_list = []
for i, nim_file in enumerate(nim_files):
    df = pd.read_csv(nim_file)
    st = 0 if i == 0 else 1
    df_list.append(df.iloc[:, st:])
df = pd.concat(df_list, axis=1)
df
```

```python
df_np = df.to_numpy()
nutrients = list(df.columns)[1:]
nims = np.array(df_np[:, 1:].T, dtype=float)

nims_df = pd.DataFrame(nims, columns=df_np[:, 0])
nims_df["nutrients"] = nutrients
nims_columns = list(nims_df.columns)
nims_columns = [nims_columns[-1]] + nims_columns[:-1]
nims_df = nims_df.reindex(columns=nims_columns)
nims_df
```

```python
nims_microbes = nims_df.columns[1:]
```

```python
patients_df = pd.read_csv("./patient_inputs.csv")
patient_microbes = patients_df.columns[3:]
```

```python
np.all(nims_microbes == patient_microbes)
```

```python
nims_df.to_csv("nims.csv", index=False)
```
